# Cleaning Data


```python
# df_c = df.copy()
```

year column have the object value like {'r 15', 'Zest', 'ara)', and another}
we need only year number not many noise letter or text

```python
df = df[df['year'].str.isnumeric()]
```

```python
df['year'].unique()
```

```python
df['year'] = df['year'].astype(int)
```

Pirce column has 'Ask For Price' and ,
price is object change it to integer

```python
df = df[df['Price']!='Ask For Price']
```

```python
df['Price'].head()
```

```python
df['Price'] = df['Price'].str.replace(',','').astype(int)
```

kms_driven column has 'Petrol', kms and , 
kms_driven split it take only integer and covet to integer

```python
df['kms_driven'] = df['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
df['kms_driven'] = df['kms_driven'].str.isnumeric().astype(int)
```

check the null value and cleanning it


```python
df = df[~df['fuel_type'].isna()]
#df.dropna()
#df.isna().sum()
```

check duplicate and drop it

```python
df.duplicated().any()
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

Company does not need any cleaning now. Changing car names. Keeping only the first three words

```python
df['name'] = df['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
df.head()
```

Resetting the index of the final cleaned data to be correct

```python
df = df.reset_index(drop=True)
```

# Model

```python
df.columns
```

Extract Training Data

```python
y = df.drop(columns=['name','company', 'year','kms_driven', 'fuel_type'])
x = df[['name','company', 'year','kms_driven', 'fuel_type']]
```

```python
print(x.shape)
print(y.shape)
```

Extract for train(80%) and test(20%)

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

create a OneHotEncoder object, One-hot encoding converts categorical data into a numerical format by creating binary columns for each category

```python
ohe = OneHotEncoder()
#ohe.fit_transform()
ohe.fit(x[['name','company', 'fuel_type']])
```

Creating a column transformer to transform categorical columns

```python
# column_tra = make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['name','company', 'fuel_type']))
```

```python
column_tra = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company', 'fuel_type']))
```

Create model linear class

```python
model = LinearRegression()
```

Make a pipeline

```python
pipe = make_pipeline(column_tra,model)
```

Apply the training and test set in model

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_tra,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```
